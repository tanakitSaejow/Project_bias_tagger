```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
model=['WangchanBERTa','Monsoon']
label=['DELETE','EDIT']
delete=[0.56,0.08]
edit=[0.07,0.05]

bar_width = 0.15
pos1 = np.arange(len(model))
pos2 = [x + bar_width for x in pos1]

plt.bar(pos1, delete, width = bar_width, color='blue', edgecolor='black')
plt.bar(pos2, edit, width = bar_width, color='pink', edgecolor='black')

plt.xticks([r + bar_width for r in range(len(model))], model)
plt.xlabel('Model', fontsize=16)
plt.ylabel('F1-Score', fontsize=16)
plt.legend(label, loc =2)
plt.savefig('sentence.png')
plt.show()
```

```python
model=['WangchanBERTa','Monsoon']
label=['B-DELETE','B-EDIT','I-DELETE','I-EDIT']
b_delete=[0.66,0.38]
b_edit=[0.07,0.05]
i_delete=[0.59,0.00]
i_edit=[0.00,0.00]

bar_width = 0.15
pos1 = np.arange(len(model))
pos2 = [x + bar_width for x in pos1]
pos3 = [x + bar_width for x in pos2]
pos4 = [x + bar_width for x in pos3]

plt.bar(pos1, b_delete, width = bar_width, color='blue', edgecolor='black')
plt.bar(pos2, b_edit, width = bar_width, color='pink', edgecolor='black')
plt.bar(pos3, i_delete, width = bar_width, color='orange', edgecolor='black')
plt.bar(pos4, i_edit, width = bar_width, color='green', edgecolor='black')

plt.xticks([r + bar_width for r in range(len(model))], model)
plt.xlabel('Model', fontsize=16)
plt.ylabel('F1-Score', fontsize=16)
plt.legend(label, loc =2)
plt.savefig('word.png')
plt.show()
```
